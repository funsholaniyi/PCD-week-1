```python
import pandas as pd
```

```python
prices = pd.read_csv('NATIONAL WHOLESALE PRICE OF SOME SELECTED COMMODITIES, 1970-2017.csv')
temperature = pd.read_csv('Monthly Average Temperature in Ghana.csv')
rainfall = pd.read_csv('RAINFALL.csv')
```

```python
prices.info()
```

```python
prices.columns = prices.columns.str.strip()
```

```python
prices.info()
```

```python
prices.head(10)
```

```python
months_map = {'JANUARY': 1, 'FEBRUARY': 2, 'MARCH': 3, 'APRIL': 4, 'MAY':5, 'JUNE':6, 'JULY':7, 'AUGUST':8, 'SEPTEMBER':9, 'OCTOBER': 10, 'NOVEMBER': 11, 'DECEMBER': 12}
```

```python
prices.MONTH = prices.MONTH.map(months_map)
```

```python
prices.head()
```

```python
prices['DATE'] = pd.to_datetime(prices[['YEAR', 'MONTH']].assign(Day=1))
```

```python
prices.head()
```

```python
prices.set_index('DATE')
```
